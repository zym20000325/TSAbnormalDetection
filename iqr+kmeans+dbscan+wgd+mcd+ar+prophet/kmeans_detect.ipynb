```python
""" 
@author: zym

"""
import numpy as np  
import pandas as pd
from sklearn.cluster import KMeans
```

```python
def Kmeans_detect(data,k):
    
    data1 = list(data.value)
    data1 = np.array(data1)
    data1 = data1.reshape(-1,1)
    
    kmeans = KMeans(n_clusters=k, random_state=8).fit(data1)
    
    data["label"]=kmeans.labels_
    
    p = list(set(kmeans.labels_))
    print("簇的数目: "+str(k)+"\n")
    
#     clusters = []
    
    for i in range(k):
        print("簇 ", p[i],":")
        one_cluster = data[data.label == i]
        print(one_cluster)
        print("共"+str(len(one_cluster))+"个点")
        print("\n")
#         clusters.append(one_cluster)
```

```python
# cpu
data = pd.read_csv("data/ec2_cpu_utilization_24ae8d.csv")
Kmeans_detect(data,6)
```

```python
# cpc
data = pd.read_csv("data/exchange-4_cpc_results.csv")
Kmeans_detect(data,6)
```

```python
# occupancy
data = pd.read_csv("data/occupancy_t4013.csv")
Kmeans_detect(data,10)
```
